```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
```

```python
model_name = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
documents = [
    "الذكاء الاصطناعي هو مجال من مجالات علوم الكمبيوتر.",
    "التعلم العميق هو جزء من الذكاء الاصطناعي.",
    "اللغة العربية هي لغة سامية."
]
```

```python
vector_db = FAISS.from_texts(documents, embeddings)
```

```python
query = "ما هو الذكاء الاصطناعي؟"
relevant_docs = vector_db.similarity_search(query, k=2)
```

```python
relevant_docs
```

```python
generator = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
context = " ".join([doc.page_content for doc in relevant_docs])
prompt = f"السؤال: {query}\nالسياق: {context}\nالإجابة:"
answer = generator(prompt, max_length=50)
print(answer[0]['generated_text'])
```
